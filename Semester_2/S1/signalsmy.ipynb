```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-12,12,1) # an evenely spaced integer array
fig, axes = plt.subplots(2,1, sharex='all',sharey='all', figsize=(10,10))
theta = 0.0

def real_of_complex_exp(c,alpha,n, omega0,theta):
    """returns an arrray of the real part of a complex exponential signal"""
    return abs(c)*(abs(alpha)**n)*np.cos(omega0*n+theta)

# create signals
re_x1n = real_of_complex_exp(1,1.1,n,2*np.pi/12,theta)
re_x2n = real_of_complex_exp(1,1.1,n,2*np.pi/12+2*np.pi,theta)

# create plots
# plot xn1
axes[0].stem(n,re_x1n)
axes[0].set_xlabel('$n$')
axes[0].set_ylabel('$Re(x_1 n)$')
axes[0].title.set_text(r'$1.1^n \cos( \frac{2 \pi}{12} n)$')
axes[0].grid(True)

# plot xn2
axes[1].stem(n,re_x2n)
axes[1].set_xlabel('$n$')
axes[1].set_ylabel('$Re(x_2 n)$')
axes[1].title.set_text(r'$1.1^n \cos( ( \frac{2 \pi}{12} + 2\pi ) n)$')
axes[1].grid(True)

plt.show()
```

when $\omega = \omega _0 + 2\pi$ the signal is the same because $\cos(\omega _0n)=\cos(\omega _0n+2\pi n)$ since $n \in \Z$

```python
fs = 1000 # sampling frequency
ts = 1/fs # sampling period
t = np.arange(-0.1,3,ts) # create array spaced evenly by sampling period

def growing_sinusoid(c, r, theta, t, omeg0):
    """returns a growing sinusoid signal array"""
    return c*np.exp(r*t)*np.cos(omeg0*t + theta)

# define values
f = 5 # frequency is 5 Hz
omeg0=2*np.pi*f # angular frquency

# create two signals
x1t = growing_sinusoid(0.15, 2, 0, t, omeg0)
x2t = growing_sinusoid(0.15, 2, 0, t, omeg0+ 2*np.pi)

# plotting the signals
fig,ax = plt.subplots(figsize=(12,6))
# plot x1t
ax.plot(t,x1t, label=r'$0.15e^{2 t} \cos (10\pi t)$')
ax.plot(t,x2t, label=r'$0.15e^{-2 t} \cos (10\pi t+2\pi t)$')
ax.set_xlabel('$t$')
ax.set_ylabel('$x(t)$')
ax.legend()
ax.grid(True)
plt.show()
```

when $\omega = \omega _0 + 2\pi$ the signal is not the same because for $\cos(\omega _0n)=\cos(\omega _0n+2\pi n)$ t should be integer $t \in \Z$. That is why when t is an integer the two signals take same x(t) value but not at any other point making it a whole different signal.

```python
n = np.arange(-12,24,1) # an evenely spaced integer array
fig, axes = plt.subplots(7,1, sharex='all',sharey='all', figsize=(12,21))
theta = 0.0

def real_of_complex_exp(c,alpha,n, omega0,theta=0):
    """returns an arrray of the real part of a complex exponential signal"""
    return abs(c)*(abs(alpha)**n)*np.cos(omega0*n+theta)

# create signals
re_x1n = real_of_complex_exp(1,1.1,n,2*np.pi/12)
re_x2n = real_of_complex_exp(1,1.1,n,2*np.pi/6)
re_x3n = real_of_complex_exp(1,1.1,n,2*np.pi/4)
re_x4n = real_of_complex_exp(1,1.1,n,np.pi)
re_x5n = real_of_complex_exp(1,1.1,n,1.2*np.pi)
re_x6n = real_of_complex_exp(1,1.1,n,1.6*np.pi)
re_x7n = real_of_complex_exp(1,1.1,n,2*np.pi)

# create plots
# plot xn1
axes[0].stem(n,re_x1n)
axes[0].set_ylabel('$Re(x_1 n)$')
axes[0].title.set_text(r'$1.1^n \cos( \frac{\pi}{6} n)$')
axes[0].grid(True)

# plot xn2
axes[1].stem(n,re_x2n)
axes[1].set_ylabel('$Re(x_2 n)$')
axes[1].title.set_text(r'$1.1^n \cos( \frac{\pi}{3} n)$')
axes[1].grid(True)

# plot xn3
axes[2].stem(n,re_x3n)
axes[2].set_ylabel('$Re(x_2 n)$')
axes[2].title.set_text(r'$1.1^n \cos( \frac{\pi}{2} n)$')
axes[2].grid(True)

# plot xn4
axes[3].stem(n,re_x4n)
axes[3].set_ylabel('$Re(x_2 n)$')
axes[3].title.set_text(r'$1.1^n \cos( \pi n)$')
axes[3].grid(True)

# plot xn5
axes[4].stem(n,re_x5n)
axes[4].set_ylabel('$Re(x_2 n)$')
axes[4].title.set_text(r'$1.1^n \cos( 1.2\pi n)$')
axes[4].grid(True)

# plot xn6
axes[5].stem(n,re_x6n)
axes[5].set_ylabel('$Re(x_2 n)$')
axes[5].title.set_text(r'$1.1^n \cos( 1.6 \pi)$')
axes[5].grid(True)

# plot xn7
axes[6].stem(n,re_x7n)
axes[6].set_ylabel('$Re(x_2 n)$')
axes[6].title.set_text(r'$1.1^n \cos( 2\pi n)$')
axes[6].grid(True)

plt.show()
```

```python
fs = 100 # sampling frequency
ts = 1/fs # sampling period
t = np.arange(-1., 1., ts) # a linearly−spaced array with step ts for time axis
fig, axes = plt.subplots(2,1, figsize=(10,10))

f = 1/2 # 2 Hz frequency for the signal
omega0 = 2*np.pi*f # angular frequency

xt = t # signal in the time domain
Xf = np.fft.fft(xt) # signal in the frequency domain
freq = np.fft.fftfreq(t.shape[-1], d=ts) # frequency axis values from the transformation

# plot the signal in time domain      
axes[0].plot(t,xt)
# axes[0].plot(t,2.*np.cos(omega0*t)) # building blocks
# axes[0].plot(t,np.cos(3*omega0*t)) # building blocks
axes[0].set_xlabel('$t$ in s')
axes[0].set_ylabel('$x(t)$')
# axes[0].title.set_text(r'$x(t)=0.75+2\cos(\omega _0t)+\cos(3\omega _0t)$')

# ranges of x_axis and y_axis for the frequency domain representation
valsubrange = np.concatenate((np.arange(0,21,1), np.arange(-1,-21,-1)))
freqsubrage = np.concatenate((np.arange(0,21,1), np.arange(-1,-21,-1)))

# plot the signal in the frequency domain
axes[1].stem(freq[freqsubrage], Xf.imag[valsubrange]/len(t)) # plot real part of freq domain signal against freq
axes[1].set_xlabel('$f$ in Hz')
axes[1].set_ylabel(r'$\mathfrak{Re}(X(j\omega))$')
plt.xticks(np.arange(-10,11))

plt.show()
# Xf
```
