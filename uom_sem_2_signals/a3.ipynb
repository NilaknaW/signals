```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftshift, ifft
from scipy import signal
```

```python
# Creating 3 sinusoidal signals
w1 = 100*np.pi
w2 = 500*np.pi
a1 = 0.75
a2 = 1
fs = 4095
ws = 2*np.pi*fs

def x(t):
    # x_t = a1*np.sin(w1*t)+ a2*np.sin(w2*t)
    x_t = np.cos(5*t)+2*np.sin(3*t)
    return x_t

time = np.linspace(-10,10,fs+1)
xt = [x(t_) for t_ in time]

Xw = fft(xt, 4096)*2*np.pi/fs
Xw = fftshift(Xw)
k = np.arange(1,4097)
w = k/4096*ws-ws/2

# plotting the input signal in the frequency domain
fig, axes = plt.subplots(2,1,figsize=(18,12))

axes[1].plot(w,abs(Xw))
axes[1].title.set_text('Signal in Frequency Domain')
axes[1].set_xlabel('$\omega$')
axes[1].set_ylabel('Magnitude')
axes[1].set_xlim(-500, 500)
# axes[1].set_xticks(np.arange(-100*np.pi, 100*np.pi+1,20*np.pi))
# axes[1].set_xticklabels([str(i)+(r'$\pi$' if i else '') for i in range(-1200,1210,100)])
# axes[1].set_xlim(-1000*np.pi, 1000*np.pi)
# axes[1].set_yticks([0,np.pi/2,np.pi])
# axes[1].set_yticklabels([0,r'$\pi$/2',r'$\pi$'])
axes[1].grid()

axes[0].plot(time,xt)
axes[0].title.set_text('Signal in Time Domain')
axes[0].set_xlabel('$t$')
axes[0].set_xlim(-10,10)
axes[0].set_ylabel('$x(t) = 0.75 \sin(100\pi t) + 1.0\sin(500\pi t)$')
axes[0].grid(True)
plt.show()
```
